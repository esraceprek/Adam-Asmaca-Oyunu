```python
import pandas as pd
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics.pairwise import linear_kernel
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import mean_absolute_error
```

```python
movies = pd.read_csv('movies.csv', usecols=['movieId', 'title' , 'genres'])
movies.head()
```

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
movies['genres'] = movies['genres'].str.split('|')
movies['genres'] = movies['genres'].fillna("").astype('str')

movies.head()



```

```python
tf = TfidfVectorizer(analyzer='word', ngram_range=(1 , 2) , min_df=0.0 , stop_words='english')

tfidf_matrix = tf.fit_transform(movies['genres'])
```

```python
cosine_similarity = linear_kernel(tfidf_matrix , tfidf_matrix)
```

```python
titles = movies['title']
indices = pd.Series(movies.index , index=movies['title'])
```

```python
def movie_recommendation(title):
  idx = indices[title]
  similarity_rate = list(enumerate(cosine_similarity[idx]))
  similarity_rate = sorted(similarity_rate , key=lambda x: x[1] , reverse=True)
  similarity_rate = similarity_rate[1:15]

  movie_indices= [i[0] for i in similarity_rate]
  return titles.iloc[movie_indices]
```
